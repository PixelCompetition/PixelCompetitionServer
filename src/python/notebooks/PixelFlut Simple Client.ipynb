```python
import socket
import numpy as np
import PIL
from PIL import Image

# server settings
# HOST = '88.99.149.14'
# HOST = 'localhost'
# HOST = '192.168.1.102'
HOST = 'box.pixel-competition.de'

PORT = 2342
length = 100;
height = 100;
maxsize = (length, height)

# connect socket
sockets = []
for i in range (0, 20):
  sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
  sock.connect((HOST, PORT))
  sockets.append(sock)

def send(str):
    for s in sockets:
        s.send(str.encode())

# prepare image (open, convert to rgba, resize, convert to array)
img = Image.open('nut.jpg')
img = img.convert('RGBA')
img.thumbnail(maxsize, PIL.Image.ANTIALIAS)
arr = np.array(img)

# set a single pixel
def setPixel(x,y,r,g,b,a=255):
  send(('PX %d %d #%02x%02x%02x\n' % (x,y,r,g,b)))
  print(('PX %d %d %d %d %d\n' % (x,y,r,g,b)))

# print image array
def image():
  for i in range(0, img.size[0]):
    for j in range(0, img.size[1]):
      setPixel(i+250, j+250, arr[j][i][0], arr[j][i][1], arr[j][i][2], 255)

image()
send ("help\n".encode())
```
